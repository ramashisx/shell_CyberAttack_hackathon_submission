### postprocessing step where it finds the part of substring from the model generated output

```python
import re

def find_substring_indices(main_string, substring):
    pattern = re.compile(re.escape(substring))
    match = pattern.search(main_string)

    if match:
        start_index = match.start()
        end_index = match.end() - 1
        return start_index, end_index
    else:
        return None
```

```python
data = pd.read_json("../data/datafinal.json")
```

```python
train_data = data[data["ContainsCode"].apply(lambda x: isinstance(x, bool))].copy()
```

```python
train_data["CodeList"] = train_data["CodeList"].apply(lambda x: x.replace('\\n', '\n').replace('\\t', '\t'))
```

### simple analysis

```python
i = 0
for _, row in train_data.iterrows():
    if row["ContainsCode"]:
        if row["CodeList"] not in row["Text"]:
            # print(row)
            i += 1
        else:
            indices = find_substring_indices(row["Text"], row["CodeList"])
            if indices is not None:
                start, end = indices
                print(row["Text"][start:end+1])
                print("------")
                # print(row["CodeList"])
                # print("-------")
                # print(row["Text"][start:end+1] == row["CodeList"])
                # print(f"Start index: {start}, End index: {end}")
            else:
                print("Substring not found in the main string.")
```
